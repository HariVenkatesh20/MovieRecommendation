```python
# Example: Extracting movie genres
genres = movie_df['genres'].str.split('|', expand=True)
genres.columns = ['genres' + str(col) for col in genres.columns]
movie_df = pd.concat([movie_df, genres], axis=1)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel


tfidf_vectorizer = TfidfVectorizer(stop_words='english')
movie_df['genres'] = movie_df['genres'].fillna('')
tfidf_matrix = tfidf_vectorizer.fit_transform(movie_df['genres'])
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

def get_recommendations(title, cosine_sim=cosine_sim):
    idx = movie_df[movie_df['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  
    movie_indices = [i[0] for i in sim_scores]
    return movie_df['title'].iloc[movie_indices]

print(get_recommendations('Inception'))
```
